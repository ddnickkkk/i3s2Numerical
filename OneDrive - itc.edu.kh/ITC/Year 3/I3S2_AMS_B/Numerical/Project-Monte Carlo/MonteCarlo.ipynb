```python
import numpy as np

def monte_carlo_simulation(initial_investment, annual_growth_rate, num_simulations, num_years):
    results = []
    
    for _ in range(num_simulations):
        investment_value = initial_investment
        
        for _ in range(num_years):
            investment_value *= (1 + np.random.normal(annual_growth_rate, 0.15))
        
        results.append(investment_value)
    
    return results

initial_investment = 100000
annual_growth_rate = 0.095
num_simulations = 10000
num_years = 30

simulation_results = monte_carlo_simulation(initial_investment, annual_growth_rate, num_simulations, num_years)

expected_value = np.mean(simulation_results)
print("Expected retirement account value:", expected_value)
```

```python
import matplotlib.pyplot as plt

# Plotting the distribution
plt.figure(figsize=(10, 6))
plt.hist(simulation_results, bins=50, density=True, alpha=0.7)
plt.xlabel('Retirement Account Value')
plt.ylabel('Probability Density')
plt.title('Monte Carlo Simulation: Retirement Account Value Distribution')

plt.show()
```
